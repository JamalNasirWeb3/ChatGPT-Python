```python
import csv
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def group_add(name):
    l=[]
    with open('/content/people.csv','rt') as f:  # Update the file path if needed
        data = csv.reader(f)
        for row in data:
            if row[1] != "Given Name":
                l.append(row[1])

    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--headless')  # Optional: Run Chrome in headless mode
    chrome_options.add_argument('--no-sandbox')  # Bypass OS security model
    chrome_options.add_argument('--disable-dev-shm-usage')  # Overcome limited resource problems

    driver = webdriver.Chrome(options=chrome_options)  # Use options parameter
    driver.maximize_window()
    driver.get('https://web.whatsapp.com/')

    # Waiting for user to scan the QR code
    input("Please scan the QR code and press Enter after you're logged in.")

    try:
        # Finding the group by its title
        group_title_xpath = f'//span[@title="{name}"]'
        group_title_element = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, group_title_xpath)))
        group_title_element.click()
    except Exception as e:
        print("WhatsApp group doesn't exist or couldn't be located.")
        print("Error:", e)
        driver.quit()
        return

    # Waiting for the chat input element to load
    chat_input_xpath = '//div[@class="_3FRCZ copyable-text selectable-text"][@data-tab="6"][@contenteditable="true"]'
    chat_input_element = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, chat_input_xpath)))

    # Clicking on the add participant button
    add_participant_button_xpath = '//div[@title="Add participants"]'
    add_participant_button = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, add_participant_button_xpath)))
    add_participant_button.click()

    # Waiting for the contact search input element to load
    contact_search_xpath = '//input[@title="Search contacts"]'
    contact_search_element = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, contact_search_xpath)))

    for contact in l:
        # Entering contact name in the search bar
        contact_search_element.clear()
        contact_search_element.send_keys(contact)
        time.sleep(2)  # Giving time for search results to load

        # Clicking on the contact
        contact_xpath = f'//span[@title="{contact}"]'
        contact_element = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, contact_xpath)))
        contact_element.click()

    # Clicking on the checkmark to confirm adding participants
    confirm_button_xpath = '//div[@title="Add"]'
    confirm_button = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, confirm_button_xpath)))
    confirm_button.click()

    driver.quit()

name = input("Enter the name of the WhatsApp group you want to add people to: ")
group_add(name)
```
